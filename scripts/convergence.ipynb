```python
def get_force_on_sphere(path_,filename):
    full_path=os.path.join(path_, filename)
    with open(full_path,'r') as f:
        lines=f.readlines()
        times=[0]*len(lines)
        force=[0]*len(lines)
        for i in range(0,len(lines)):
            cur_line=lines[i].split(' ')
            times[i]=float(cur_line[0])
            force[i]=float(cur_line[1].split('\n')[0])
    return times,force
```

```python
#initialize globals
path_to_sphere_con='/home/clove/rheoTool/of90/custom_models/sphere_con/Oldroyd-BLog'
path_to_sphere_con_sys=path_to_sphere_con+'/system'
path_to_sphere_con_tfin=path_to_sphere_con+'/0.3'
base_cells=(np.linspace(1,19,num=9,endpoint=True))
run_times=np.zeros(len(base_cells))
```

```python
#prepare true value
true_base=20
construct_blockmesh(path_to_sphere_con_sys,true_base)
os.system(f"gnome-terminal -e 'bash -c \"cd ..;cd ..;cd {path_to_sphere_con};./Allclean;./Allrun; exit; exec bash\"'")
start=time.time()
end=start
while os.path.exists(path_to_sphere_con_tfin)==False:
    end=time.time()
print(end-start)
true_times, true_force = get_force_on_sphere(path_to_sphere_con,'sphere_force.txt')
```

```python
errors=np.zeros(len(base_cells))
for i in range(0,len(base_cells)):
    construct_blockmesh(path_to_sphere_con_sys,1)
    os.system(f"gnome-terminal -e 'bash -c \"cd ..;cd ..;cd {path_to_sphere_con};./Allclean;./Allrun; exit; exec bash\"'")
    start=time.time()
    end=start
    while os.path.exists(path_to_sphere_con_tfin)==False:
        end=time.time()
    run_times[i]=(end-start)
    times,forces=get_force_on_sphere(path_to_sphere_con,'sphere_force.txt')
    for x in range(0,len(forces)):
        if true_force[x]!=0:
            errors[i]+=abs(forces[x]-true_force[x])/abs(true_force[x])
```
